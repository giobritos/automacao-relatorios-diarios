Fazer a importação das bibliotecas que nos ajudarão no processo de automação.

```python
# a palavra import indica ao python que precisamos de uma determinada biblioteca
import pyautogui
import time
import pyperclip
```

Nossa próxima etapa é fazer o download da base de dados. 

O nosso objetivo aqui é não utilizar o mouse ou teclado para realizar as seguintes tarefas:

1) Abrir um navegador de internet;

2) Copiar o link do Google Drive onde está nossa base de dados;

3) Entrar no Google Drive onde está nossa base de dados;

4) Abrir pasta Aula 1 do Google Drive;

5) Baixar planilha.

```python
# como já temos um navegador aberto por usar o Jupyter abriremos apenas uma nova guia
pyautogui.PAUSE = 1
pyautogui.alert("Vai começar, aperte OK e não mexa em nada!")
pyautogui.hotkey('ctrl', 't')

# aqui abriremos o drive onde está nossa base de dados, e esperaremos o site abrir
pyperclip.copy("https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing")
pyautogui.hotkey("ctrl", "v")
pyautogui.press("enter")

time.sleep(15)

# baixando a base de dados atulizada
pyautogui.click(x=442, y=292, clicks=2)
time.sleep(2)
pyautogui.click(x=428, y=365)
pyautogui.click(x=1713, y=190)
pyautogui.click(x=1520, y=597)
time.sleep(5)
```

Agora iremos iniciar a nossa análise com o uso da biblioteca pandas.

```python
import pandas as pd

df = pd.read_excel(r'C:\Users\Giovana\Downloads\Vendas - Dez.xlsx')
display(df)
faturamento = df['Valor Final'].sum()
qtde_produtos = df['Quantidade'].sum()
```

Por fim, iremos criar um relatório e enviá-lo via e-mail. Os passos que precisamos seguir são: 

1) ABRIR o gmail;

2) Clicar em ESCREVER (novo e-mail); 

3) Escrever o DESTINATÁRIO; 

4) Selecionar o campo ASSUNTO;

5) Selecionar o campo CORPO DO E MAIL e escrever o e mail usando os indicadores calculados;

6) ENVIAR E mail.

```python
# entrando no gmail
pyautogui.hotkey("ctrl", "t")
pyperclip.copy("https://mail.google.com/mail/u/0/#inbox")
pyautogui.hotkey("ctrl", "v")
pyautogui.press("enter")
time.sleep(5)

# enviando o e-mail
pyautogui.click(x=93, y=210)
time.sleep(7)

# adiciona os emails
pyautogui.write("diretoriaxxxxxxxx@gmail.com")
pyautogui.press("tab")
pyautogui.write("financeiroxxxxxxxx@gmail.com")
pyautogui.press("tab")

# pula pro campo de assunto
pyautogui.press("tab") 

# escreve o assunto
pyperclip.copy("Relatório de Vendas")
pyautogui.hotkey("ctrl", "v") 

# pula pro corpo do email
pyautogui.press("tab") 

# escreve o texto
texto = f"""
Prezados, bom dia

O faturamento de ontem foi de: R${faturamento:,.2f}
A quantidade de produtos foi de: {qtde_produtos:,}

Atenciosamente,
Giovana"""

pyperclip.copy(texto)
pyautogui.hotkey("ctrl", "v")

# clica no botão enviar
pyautogui.hotkey("ctrl", "enter")
```
